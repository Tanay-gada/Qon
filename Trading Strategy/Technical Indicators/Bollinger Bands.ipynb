```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np


import warnings
warnings.filterwarnings("ignore")

# to fetch data 
import yfinance as yf

import talib as ta
```

### Import stock data of TATA Motors

```python
symbol = 'TATAMOTORS.NS'
start = '2020-01-01'
end = '2021-01-01'

# Read data 
df = yf.download(symbol,start,end)

df.head()
```

```python
n = 20
SMA = df['Adj Close'].rolling(n).mean()
STD = df['Adj Close'].rolling(n).std()
df['up_band'] = SMA + 2*STD
df['low_band'] = SMA - 2*STD
df['mid_band']= SMA
```

```python
#direct from lib function
#df['up_band'], df['mid_band'], df['low_band'] = ta.BBANDS(df['Close'], timeperiod =20)
```

```python
plt.figure(figsize=(20,7))
labels = ['Close','up_band', 'mid_band', 'low_band']
plt.plot(df[['Close','up_band', 'mid_band', 'low_band']], label=labels)
plt.xlabel('Date')
plt.legend()
plt.show()
```
