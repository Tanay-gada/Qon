```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np


import warnings
warnings.filterwarnings("ignore")

# to fetch data 
import yfinance as yf

import talib as ta
```

### Import stock data of TATA Motors

```python
symbol = 'TATAMOTORS.NS'
start = '2020-01-01'
end = '2021-01-01'

# Read data 
df = yf.download(symbol,start,end)

df.head()
```

```python
n = 20
df['SMA'] = df['Adj Close'].rolling(n).mean()
```

```python
plt.figure(figsize=(20,7))
labels = ['Adj Close','SMA']
plt.plot(df[['Adj Close','SMA']], label=labels)
plt.xlabel('Date')
plt.legend()
plt.show()
```

## Exponential Moving Average

```python
df['EMA'] = ta.EMA(df['Adj Close'], timeperiod = 20)
```

```python
plt.figure(figsize=(20,7))
labels = ['Adj Close','EMA']
plt.plot(df[['Adj Close','EMA']], label=labels)
plt.xlabel('Date')
plt.legend()
plt.show()
```
